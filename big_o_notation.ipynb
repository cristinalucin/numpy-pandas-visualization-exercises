```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

### Curves to graph

     - y = 0n + 1 and label the curve "O(1)
     - y = log(n) and label the curve "O(log n)
    - y = n and label the curve "O(n)
    - y = n*log(n) and label it "O(n log n)
    - y = n ** 2 and label it "O(n^2)"
    - y = 2^n and label it "O(2^n)"
    - y = n! and label it "O(n!)"
    - y = n^n and label it "O(n^n)"

```python
# y = 0n + 1 and label the curve "O(1)

x = list(range(1, 20))
y = [(0 * n + 1) for n in x]

plt.scatter(x, y, c='blue')
plt.plot(x, y, c='blue', label='$0n + 1$')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='upper right')

plt.show()
```

```python
# y = log(n) and label the curve "O(log n)

x = list(range(1, 20))
y = [np.log(n) for n in x]

plt.scatter(x, y, c='red')
plt.plot(x, y, c='red', label='$y=log(n)$')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='lower right')

plt.show()
```

```python
# y = n and label the curve "O(n)"

x = list(range(1, 20))
y = [n for n in x]

plt.scatter(x, y, c='green')
plt.plot(x, y, c='green', label='$y = x$')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='lower right')

plt.show()
```

```python
# y = n*log(n) and label it "O(n log n)

x = list(range(1, 20))
y = [(n * np.log(n)) for n in x]

plt.scatter(x, y, c='orange')
plt.plot(x, y, c='orange', label='$y = n*log(n)$')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='lower right')

plt.show()
```

```python
#y = n ** 2 and label it "O(n^2)"

x = list(range(1, 20))
y = [(n ** 2) for n in x]

plt.scatter(x, y, c='purple')
plt.plot(x, y, c='purple', label='$y = n ** 2$')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='lower right')

plt.show()
```

```python
# y = 2^n and label it "O(2^n)"

x = list(range(1, 20))
y = [ 2 ** n for n in x]

plt.scatter(x, y, c='pink')
plt.plot(x, y, c='pink', label='$y = 2 ^ n $')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='upper left')

plt.show()
```

```python
# y = n! and label it "O(n!)"

x = list(range(1, 10))
y = [ math.factorial(n) for n in x]

plt.scatter(x, y, c='olive')
plt.plot(x, y, c='olive', label='$y = n!$')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='lower right')

plt.show()
```

```python
# y = n^n and label it "O(n^n)"

x = list(range(1, 5))
y = [n ** n for n in x]

plt.scatter(x, y, c='cyan')
plt.plot(x, y, c='cyan', label='$y = n ** n$')

plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend(loc='upper left')

plt.show()
```
